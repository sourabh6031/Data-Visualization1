```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('games.csv')
data
```

```python
plt.figure(figsize=(11,10))
sns.scatterplot(data=data,x='NA_Sales',y='EU_Sales')
```

```python
# filtering multiple top values
```

```python
top_genre = data['Genre'].value_counts().index[:3]
top_publisher = data['Publisher'].value_counts().index[:3]
top_platform = data['Platform'].value_counts().index[:3]
```

```python
top_3_data = data.loc[(data['Genre'].isin(top_genre)) & (data['Publisher'].isin(top_publisher)) & (data['Platform'].isin(top_platform))] 
top_3_data
```

### ASSIGNMENT

```python
# Q1. From the Titanic dataset what can you infer about the impact of ages of passengers and the information about 
#     if they were alone on the chances of their survival?
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=titanic, x='alive', y='age', hue='alone')
plt.grid()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=titanic, x='alone', y='age', hue='alive')
plt.grid()
```

```python
# CONCLUSION
# People who were alone and survived were found to have a bit higher median age than those who did not survive.
# People who were not alone and survived were found to have lower median age than those who did not survive.
```

```python
# Q2. For the Titanic dataset, what can you say about the impact of fare on the survival dependency differently for all the classes?
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=titanic, x='class', y='fare', hue='survived',estimator=np.sum)
plt.grid()
```

```python
titanic.loc[titanic['class']=='Third'].describe()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=titanic, x='class', y='fare', hue='survived',estimator=np.mean)
plt.grid()
```

```python
# for option 4
```

```python
second = titanic.loc[(titanic['class']=='Second') & (titanic['fare']>60)]
second
```

```python
sns.barplot(data = second, x='class',y='fare',hue='alive')
plt.grid()
```

```python
# CONCLUSION
# It is observed that the number of passengers from class "Third" who survived was way lesser than those who didn't survive in that class.
# In the class "Second", it is found that people who paid higher amount of fare (i.e. >60) haven't survived dominantly in that fare range.
```
